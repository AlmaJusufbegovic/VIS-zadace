# **3) Chi-square test nezavisnosti kategorijskih varijabli**

 **3.1)** KoristeÄ‡i chi-square test za ispitivanje nezavisnosti kategorijskih varijabli prvo Ä‡emo ispitati kakva je zavisnost izmeÄ‘u arome kafe i ukusa kafe.
Prvobitno vaÅ¾no je postaviti nultu **(H0)** i alternativnu hipotezu **(H1)** koje u ovom sluÄaju izgledaju ovako:


**H0** - Aroma kafe je nezavisna od njenog ukusa


**H1** - Aroma kafe zavisi od ukusa kafe

Potrebno je odrediti **nivo znaÄajnosti** na osnovu kojeg Ä‡emo razmatrati prihvatanje/odbacivanje nul-hipoteze.

```python
import pandas as pd
from scipy.stats import chi2
data=pd.read_csv('coffee/df_arabica_clean.csv')
R = data['Aroma'].nunique()
C = data['Acidity'].nunique()
# Stepen slobode
df = (R - 1) * (C - 1)
# Nivo znaÄajnosti (Î±)
alpha = 0.05
kriticnavr = chi2.ppf(1 - alpha, df)
print("Nivo znaÄajnosti: ",alpha)
print("KritiÄna vrijednost: ", kriticnavr)
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

data=pd.read_csv('coffee/df_arabica_clean.csv')

contingency_table = pd.crosstab(data['Aroma'], data['Flavor'])

# chi-square test
chi2, p_value, _, _ = chi2_contingency(contingency_table)

print("Rezultati testa:")
print("Chi-square statistika:", chi2)
print("p-vrijednost:", p_value)
```

**ANALIZA:  Ova analiza ukazuje na to da postoji veoma vaÅ¾na statistiÄki znaÄajna zavisnost izmeÄ‘u arome kafe i ukusa kafe. BuduÄ‡i da je p-vrijednost izuzetno mala, mnogo manja od naÅ¡eg nivoa znaÄajnosti od 0.05 (5%), imamo snaÅ¾ne dokaze da se nulta hipoteza (H0) moÅ¾e odbaciti**.

Ovi rezultati ukazuju na to da je aroma kafe povezana sa ukusom kafe, Å¡to je oÄekivano s obzirom na prirodu kafe i naÄin na koji aroma moÅ¾e uticati na ukus

3.5 Chi- Square test nezavisnosti varijabli Metode obrade kafe i njene arome.

Prvo postavimo hipoteze:

H0 : Aftertaste je nezavisan od naÄina obrade kafe

H1 : Aftertaste zavisi od naÄina obrade kafe



```python
import pandas as pd
from scipy.stats import chi2, chi2_contingency

data = pd.read_csv('coffee/df_arabica_clean.csv')

R = data['Processing Method'].nunique()
C = data['Aftertaste'].nunique()


df = (R - 1) * (C - 1)


alpha = 0.05


kriticnavr = chi2.ppf(1 - alpha, df)

print("Nivo znaÄajnosti:", alpha)
print("KritiÄna vrijednost:", kriticnavr)

# IzraÄunavanje kontingencijske tablice
contingency_table = pd.crosstab(data['Processing Method'], data['Aftertaste'])

# Chi-square test
chi2, p_value, _, _ = chi2_contingency(contingency_table)

print("\nRezultati testa:")
print("Chi-square statistika:", chi2)
print("p-vrijednost:", p_value)
```

S obzirom na visoku vrijednost Chi-square statistike i izuzetno nisku p-vrijednost, moÅ¾emo zakljuÄiti da postoji statistiÄki znaÄajna veza izmeÄ‘u nacina pripreme kafe (Processing Method) i arome (Aftertaste).

Odbacujemo nultu hipotezu da su nacin pripreme kafe i aroma nezavisni. Umjesto toga, prihvaÄ‡amo alternativnu hipotezu da postoji znaÄajna veza izmeÄ‘u nacina pripreme kafe i aftertaste-a. Ovi rezultati ukazuju da nacin pripreme kafe moÅ¾e imati utjecaj na njezin aftertaste.

VaÅ¾no je napomenuti da je p-vrijednost vrlo blizu nule, Å¡to ukazuje na vrlo visoku statistiÄku znaÄajnost rezultata. To sugerira da je vjerojatnost dobivanja takvih ili ekstremnijih rezultata sluÄajnim uzorkom izuzetno mala.

Ukratko, temeljem dobivenih rezultata, moÅ¾emo zakljuÄiti da postoji statistiÄki znaÄajna veza izmeÄ‘u nacina pripreme kafe i aftertaste-a.

Postavljamo hipoteze:
Nulta hipoteza (H0): Komapnije i ukus kafe su nezavisne, tj. nema znaÄajne veze izmeÄ‘u kompanija i ukusa kave.
Alternativna hipoteza (H1): Kompanije i ukus kafe su povezane, tj. postoji znaÄajna veza izmeÄ‘u kompanija i ukusa kave.

```python
import pandas as pd
from scipy.stats import chi2_contingency

# UÄitavanje CSV datoteke
df = pd.read_csv('coffee/df_arabica_clean.csv')

# Stvaranje kontingencijske tablice
contingency_table = pd.crosstab(df['Company'], df['Aftertaste'])

# Chi-kvadrat test nezavisnosti
chi2, p_value, _, _ = chi2_contingency(contingency_table)

# Ispis rezultata testa
print("Vrijednost chi-kvadrat: ", chi2)
print("P-vrijednost: ", p_value)

# Interpretacija rezultata
alpha = 0.05  # Razina znaÄajnosti
if p_value < alpha:
    print("Odbacujemo nultu hipotezu.")
    print("Postoji znaÄajna veza izmeÄ‘u kompanija i ukusa kave.")
else:
    print("PrihvaÄ‡amo nultu hipotezu.")
    print("Nema dovoljno dokaza za zakljuÄiti da postoji veza izmeÄ‘u kompanija i ukusa kave.")
```

Stoga, na temelju ovog testa, ne moÅ¾emo zakljuÄiti da postoji statistiÄki znaÄajna veza izmeÄ‘u kompanija i ukusa kave. Nulta hipoteza se prihvata, Å¡to znaÄi da nema dovoljno dokaza za tvrdnju da postoji veza izmeÄ‘u kompanija i ukusa kafe.



#**4) Provjera da li numeriÄke varijable prate normalnu distribuciju**


Za provjeru normalne distribucije koristit Ä‡emo Kolmogorov-Smirnov i Shapiro-Wilk statistiÄke testove, i QQ-plot kao grafiÄki test.

```python
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt

data=pd.read_csv('coffee/df_arabica_clean.csv')
numeric_variables = ['Number of Bags']

for variable in numeric_variables:
    variable_data = data[variable]

    ks_statistic, ks_p_value = stats.kstest(variable_data, 'norm')
    print(f"Kolmogorov-Smirnov test za {variable}: \nstatistiku = {ks_statistic}, \np-vrijednost = {ks_p_value}")
    shapiro_statistic, shapiro_p_value = stats.shapiro(variable_data)
    print(f"\nShapiro-Wilk test za {variable}: \nstatistiku = {shapiro_statistic}, \np-vrijednost = {shapiro_p_value}")

    # QQ-plot
    stats.probplot(variable_data, dist="norm", plot=plt)
    plt.title(f"QQ-plot za {variable}")
    plt.show()
```

**4.1) ANALIZA: Na temelju datih rezultata Kolmogorov-Smirnov testa i Shapiro-Wilk testa za varijablu Number of Bags (broj pakovanja) moÅ¾emo zakljuÄiti da varijabla ne slijedi normalnu distribuciju. Oba testa su dala veoma male p-vrijednosti, a i vizualnom procjenom QQ-plota vidi se znatno odstupanje podataka od linije normalne distribucije.**

# **8) Linearna regresija**

Linearna regresija je statistiÄka metoda koja se koristi za modeliranje linearnog odnosa izmeÄ‘u nezavisne varijable (poznate i kao prediktivna ili objaÅ¡njena varijabla) i jedne ili viÅ¡e zavisnih varijabli (poznate i kao kriterijske varijable). Linearna regresija zahtijeva da veza izmeÄ‘u nezavisnih varijabli i zavisne varijable bude linearna. TakoÄ‘er, preduvjet homoskedastiÄnosti zahtijeva konstantnu varijancu reziduala.

**8.1)**Na osonovu varijabli: kiselost, ukus i aftertaste Ä‡emo pokuÅ¡ati napraviti model linearne regresije za ocjenu kafe.

Prvo Ä‡emo provjeriti da li moÅ¾emo postaviti model linearne regresije za ovaj set varijabli. Da bismo to uradili, koristi Ä‡emo **scatter plot**, **matricu korelacije** i obavit Ä‡emo **test homoskedastiÄnosti** (homoskedastiÄnost se odnosi na pretpostavku linearnog regresijskog modela koja implicira da varijabilnost pogreÅ¡ke, odnosno reziduala modela ostaje konstantna za sve vrijednosti nezavisnih varijabli).

```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv('coffee/df_arabica_clean.csv')
X = data['Acidity']
y = data['Total Cup Points']
plt.scatter(X, y, color='lightgreen')
plt.xlabel('kiselost')
plt.ylabel('Ocjena kafe')
plt.title('Scatter plot za kiselost i ocjenu')
plt.show()
```

```python
import pandas as pd

print("Matrica korelacije za kiselost i ocjenu kafe")
data = pd.concat([X, y], axis=1)
correlation_matrix = data.corr()
print(correlation_matrix)
```

```python
import statsmodels.api as sm
from statsmodels.compat import lzip

print("Test homoskedastiÄnosti za kiselost i ocjenu kafe:")
X_with_constant = sm.add_constant(X)
model = sm.OLS(y, X_with_constant)
results = model.fit()
name = ['Lagrange multiplier statistic', 'p-value', 'f-value', 'f p-value']
test = sm.stats.diagnostic.het_breuschpagan(results.resid, X_with_constant)
print(lzip(name, test))
```
