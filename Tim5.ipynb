# Da li postoji povezanost između Državnog ulaganja u zdravstvo i prosječnog životnog vijeka konkretno za Bosnu i Hercegovinu

```python
import pandas as pd
import matplotlib.pyplot as plt
!git clone https://github.com/Senka2112/VIS2023-datasets


df = pd.read_csv('VIS2023-datasets/LifeExpectancyData.csv')


df_bih = df[df['Country'] == 'Bosnia and Herzegovina']


df_bih_filtered = pd.DataFrame(columns=['Year', 'Total expenditure', 'Life expectancy '])
for godina in range(2000, 2016):
    df_godina = df_bih[(df_bih['Year'] == godina)]
    df_godina_filtered = df_godina.loc[:, ['Year', 'Total expenditure', 'Life expectancy ']]
    df_bih_filtered = pd.concat([df_bih_filtered, df_godina_filtered])

# Kreiranje grafika za sve godine
fig, ax = plt.subplots()
for godina in range(2000, 2016):
    df_year = df_bih_filtered[df_bih_filtered['Year'] == godina]
    ax.scatter(df_year['Total expenditure'], df_year['Life expectancy '], label=godina)
    for i, red in df_year.iterrows():
        ax.annotate(godina, (red['Total expenditure'], red['Life expectancy ']))

ax.set_title('Odnos između ulaganja u zdravstvo i životnog vijeka za Bosnu i Hercegovinu')
ax.set_xlabel('Uaganje u zdravstvo')
ax.set_ylabel('Životni vijek')
ax.legend()
plt.show()
```

Iz priložene vizualizacije podataka možemo zaključiti iduće:
Scatter plot jasno pokazuje zavisnost između ulaganja u zdravstvo i očekivanog životnog vijeka.
Konkretno na primjeru Bosne i Hercegovine u periodu između 2000 i 2015 godine jasno se uočava da je životni vijek imao najmanje vrijednosti u godinama 2000,2003 i 2005 upravo kada je procenat izdvajanja za zdravstvo bio ispod 8.5%. Slično najveće vrijednosti ocčekivanog životnog vijeka izmjerene su u godinama 2011,2012,2013 i 2014 u kojima je također bila najveća vrijednost ulaganja u zdravstvo, tj. iznad 9.5%.
Izdvajanje u zdravstvo nije jedini faktor koji utiče na životni vijek što se najbolje očituje upoređuju vrijednosti iz 2004 i 2006. Iako je u 2004 ulaganje u zdravstvo bilo za čak 10% više u odnosu na 2006, životni vijek je bio veći u godini 2006.
To pokazuje da iako ulaganje u zdravstvo ima veliki utjecaj u očekivanom životnom vijeku, ono nije jedini faktor koji utiče na tu vrijednost.

# Proračun vjerovatnoće smrti od HiV-a

U datom .csv file-u imamo podatke o broju smrti uzrokavanih HiV/AIDS-om u životnom periodu do 4 godine, kao i broj smrti na 1000 stanovnika u periodu do 5 godina. Izračunati ćemo vjerovatnoću da je smrt u periodu do 5 godina uzrokovana HiV/AIDS-om u razvijenim zemljama.

```python
import pandas as pd


df = pd.read_csv("VIS2023-datasets/LifeExpectancyData.csv")


razvijene = df[(df["Status"] == "Developed") & (df["Year"].between(2000, 2005))]

# Izračunavanje broja smrti do 5 godina uzrokovanih HIV-om
hiv_smrti = razvijene[" HIV/AIDS"].sum()

# Izračunavanje ukupnog broja smrti do 5 godina
sve_smrti = razvijene["under-five deaths "].sum()

# Izračunavanje vjerojatnosti da će smrt do 5 godina biti uzrokovana HIV-om
if sve_smrti == 0:
    vjerovatnoca = 0
else:
    vjerovatnoca= hiv_smrti / sve_smrti

# Ispis rezultata
print("Vjerojatnoća smrti do 5 godina uzrokovane HIV-om u razvijenim zemljama u razdoblju od 2000-2005 je:", "{:.6f}".format(vjerovatnoca))
```

Primjećujemo da je vjerovatnoća da je uzrok te smrti izrazito nizak te skoro nepostojeći. Razlog tome može biti što smo koristili podatke za razvijene zemlje. Pogledajmo sada šta kažu podaci za nerazvijene zemlje.

```python
import pandas as pd


df = pd.read_csv("VIS2023-datasets/LifeExpectancyData.csv")


nerazvijene = df[(df["Status"] == "Developing") & (df["Year"].between(2000, 2005))]

# Izračunavanje broja smrti do 5 godina uzrokovanih HIV-om
hiv_smrti = nerazvijene[" HIV/AIDS"].sum()

# Izračunavanje ukupnog broja smrti do 5 godina
sve_smrti = nerazvijene["under-five deaths "].sum()

# Izračunavanje vjerojatnosti da će smrt do 5 godina biti uzrokovana HIV-om
if sve_smrti == 0:
    vjerovatnoca = 0
else:
    vjerovatnoca= hiv_smrti / sve_smrti

# Ispis rezultata
print("Vjerojatnoća smrti do 5 godina uzrokovane HIV-om u nerazvijenim zemljama u razdoblju od 2000-2005 je:", "{:.6f}".format(vjerovatnoca))
```

Vjerovatnoća smrti uzrokovane HIV/AIDS-om za nerazvijene zemlje je gotovo identičan rezultatu dobijenom za razvijene zemlje. Moguć razlog tome jeste što iako je virus HiV mnogo češći u neazvijenim zemljama, to nije slučaj za sve nerazvijene zemlje te kako nam je uzorak veliki, vrijednosti su ujednačene.
Posmatrajmo sada rezultate u konkretnom primjeru za afričku državu Boswanu.

```python
import pandas as pd


df = pd.read_csv("VIS2023-datasets/LifeExpectancyData.csv")

botswana = df[(df["Country"] == "Botswana") & (df["Year"].between(2000, 2005))]

# Izračunavanje broja smrti do 5 godina uzrokovanih HIV-om
hiv_smrti = botswana[" HIV/AIDS"].sum()

# Izračunavanje ukupnog broja smrti do 5 godina
sve_smrti = botswana["under-five deaths "].sum()

# Izračunavanje vjerojatnosti da će smrt do 5 godina biti uzrokovana HIV-om
if sve_smrti == 0:
    hiv_prob = 0
else:
    hiv_prob = hiv_smrti / sve_smrti

# Ispis rezultata
print("Vjerojatnoća smrti do 5 godina uzrokovane HIV-om u Botswani u razdoblju od 2000-2005 je:", "{:.6f}".format(hiv_prob))
```

Rezultat istraživanja za Botswanu znatno odstupa od vrijednosti koje smo dobili za razvijene i nerazvijene zemlje. Stopa umrlosti od HiV-a kod djece u razdoblju od 0 do 5 godina je čak za 166 puta veća u odnosu na ranije dobijene rezultate. Slične rezultate kao i za Botswanu dobijamo i kada umjesto nje stavimo neke druge afričke zemlje kao Lesotho, Malawi, CentralnuAfričku Republiku itd.
Iz svega ovoga možemo zaključiti da zaraženost i smrtnost HiV virusa u periodu du 5 godina ne zavisi od razvijenosti zemlje. Na globalnom nivou oko 0.05% smrti kod djece u periodu do 5 godina je izazvano HiV virusom. Međutim znatan ekstrem ovih vrijednosti dobijamo u zemljama Afričkog kontinenta. Vrijednosti na tom kontinentu idu čak i do 10% što je znatno povećanje u odnosu na prosjek te se može zaključiti da je HiV značajan problem u Africi.

# Kretanje životnog vijeka razvijenih i zemalja u razvoju

U ovom dijelu ćemo analizirati kako se prosječni životni vijek mijenjao od 2000 do 2015 godine za razvijene zemlje i zemlje u razvoju.


```python
import pandas as pd
import matplotlib.pyplot as plt
!git clone https://github.com/Senka2112/VIS2023-datasets


data = pd.read_csv('VIS2023-datasets/LifeExpectancyData.csv')

razvijene = data[data['Status'] == 'Developed']
nerazvijene = data[data['Status'] == 'Developing']

godine = list(range(2000, 2016))
razvijene_2000_2015 = razvijene.loc[razvijene['Year'].isin(godine)].groupby('Year')['Life expectancy '].mean()
nerazvijene_2000_2015 = nerazvijene.loc[nerazvijene['Year'].isin(godine)].groupby('Year')['Life expectancy '].mean()

plt.plot(razvijene_2000_2015.index, razvijene_2000_2015.values, label='Razvijene zemlje')

plt.plot(nerazvijene_2000_2015.index, nerazvijene_2000_2015.values, label='Zemlje u razvoju')

plt.title('Promjena prosječne životne dobi od 2000. do 2015. godine')
plt.xlabel('Godina')
plt.ylabel('Prosječna životna dob')

plt.legend()

plt.show()
```

Primjetimo da se značajan porast desio u oba slučaja. Sa prikaza također možemo očitati da postoji značajna razlika u životnom vijeku između razvijenih zemalja i zemalja u razvoju, cca 10 godina razlike.
Ukupan porast u očekivanom životnom vijeku ćemo izračunati u idućem dijelu koda.

Iako se čini neznatno, zemlje u razvoju u razdobllju od 15 godina imaju za 20% veći porast očekivanog životnog vijeka.

### Odnos očekivanog trajanja života i prihoda po osobi (na godišnjoj razini) za pet europskih zemalja

```python
import pandas as pd
import matplotlib.pyplot as plt
!git clone https://github.com/Senka2112/VIS2023-datasets

# Učitavanje podataka
data = pd.read_csv('VIS2023-datasets/LifeExpectancyData.csv')

# Odabir pet europskih zemalja
countries = ['Germany', 'France', 'Italy', 'Spain', 'Sweden']
data = data[data['Country'].isin(countries)]

# Line plot očekivanog trajanja života i prihoda po osobi za pet zemalja
fig, ax1 = plt.subplots(figsize=(10,5))
ax2 = ax1.twinx()

for country in countries:
    country_data = data[data['Country'] == country]
    ax1.plot(country_data['Year'], country_data['Life expectancy '], label=country)
    ax2.plot(country_data['Year'], country_data['Income composition of resources'], label=country, linestyle='--')

# Postavke osi
ax1.set_xlabel('Godina')
ax1.set_ylabel('Očekivano trajanje života')
ax2.set_ylabel('Prihod po osobi')

# Postavke grafa
ax1.legend(loc='center left', bbox_to_anchor=(1.2, 0.5))
ax2.legend(loc='center left', bbox_to_anchor=(1.2, 0.7))
plt.title('Kretanje očekivanog trajanja života i prihoda po osobi za pet europskih zemalja')
plt.show()
```

Ovdje možemo vidjeti grafik kretanja očekivanog trajanja života i prihoda po osobi na godišnjoj razini za pet europskih zemalja (Njemačka, Francuska, Italija, Španjolska i Švedska) u periodu od 2000. do 2015. godine.

Isprekidane linije predstavljaju kretanje prihoda po osobi u pet zemalja, dok pune linije predstavljaju kretanje očekivanog trajanja života u istim pet zemalja.

Grafik ukazuje na to da su sve pet zemalja bilježile porast očekivanog trajanja života i prihoda po osobi tijekom istog perioda, ali nisu imale jednak rast. Švedska je zabilježila najveći rast očekivanog trajanja života i prihoda po osobi u usporedbi s ostalim zemljama, dok je Italija zabilježila najmanji rast.

Također, grafik pokazuje pozitivnu korelaciju između prihoda po osobi i očekivanog trajanja života. To sugerira da ljudi u bogatijim zemljama imaju veću vjerojatnost da će doživjeti dulji životni vijek.

Također ćemo navesti i promjenu prosječnog prihoda po glavi stanovnika u navedenim državama u periodu od 2000. do 2015. godine.

```python
import pandas as pd
import numpy as np
!git clone https://github.com/Senka2112/VIS2023-datasets

# Učitavanje podataka
data = pd.read_csv('VIS2023-datasets/LifeExpectancyData.csv')

# Odabir pet europskih zemalja
countries = ['Germany', 'France', 'Italy', 'Spain', 'Sweden']
data = data[data['Country'].isin(countries)]

# Računanje prosječnog prihoda po osobi za svaku godinu i zemlju
avg_income = data.groupby(['Country', 'Year'])['Income composition of resources'].mean().reset_index()

# Računanje promjene prosječnog prihoda po osobi za svaku zemlju
for country in countries:
    start_income = avg_income[(avg_income['Country'] == country) & (avg_income['Year'] == 2000)]['Income composition of resources'].values[0]
    end_income = avg_income[(avg_income['Country'] == country) & (avg_income['Year'] == 2015)]['Income composition of resources'].values[0]
    change = (end_income - start_income) / start_income * 100
    print(f"Promjena prosječnog prihoda po osobi u {country} u razdoblju od 2000. do 2015. godine iznosi {change:.2f}%.")
```
